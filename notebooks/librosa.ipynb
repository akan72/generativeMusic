```python
from __future__ import print_function, division

from scipy.io import wavfile
import pandas as pd
import numpy as np 
import librosa
import pylab
import librosa.display
import sys
import os

sys.path.append('../scripts')

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
y, sr = librosa.load('../data/audio/01_10003_1-04_Shri_Visvanatham.wav',
                     offset=25.0,
                     duration=20.0)

spectrogram = np.abs(librosa.stft(y))
melspec = librosa.feature.melspectrogram(y=y,
                                         sr=sr)
chroma = librosa.feature.chroma_cqt(y=y,
                                    sr=sr)
tonnetz = librosa.feature.tonnetz(y=y, sr=sr)
```

```python
plt.figure()
plt.plot(y)
```

```python
pylab.axis('off') # no axis
pylab.axes([0., 0., 1., 1.], frameon=False, xticks=[], yticks=[]) # Remove the white edge
S = librosa.feature.melspectrogram(y=y, sr=sr)
librosa.display.specshow(librosa.power_to_db(S, ref=np.max))
pylab.savefig('../figures/spectrogram.jpg', bbox_inches=None, pad_inches=0)
pylab.show()
pylab.close()
```
